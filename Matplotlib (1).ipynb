```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Iris_new (1).csv')
```

```python
df
```

### A. Creating and customizing line charts using Matplotlib

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Id'], df['SepalLengthCm'], marker='^', linestyle='-', color='b')
plt.title("Sepal Length Across Iris Records")
plt.xlabel("Record Id")
plt.ylabel("Sepal Length (cm)")
plt.grid(True)

plt.show()
```

### B. Visualizing relationships between two or more variables using scatter plots

```python
df["color_by_species"] = df["Species"].map(lambda x: "r" if x=="Iris-setosa" else "b" if x=="Iris-virginica" else "y")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'], c=df["color_by_species"], marker='^')
plt.title("Sepal Length vs. Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True)

plt.show()
```

### C. Studying distributions of variables using histograms & bar charts to


```python
plt.figure(figsize=(10, 6))
plt.hist(df['SepalLengthCm'], bins = 15 , color = 'g', edgecolor = 'r')
plt.title("distribution of sepal length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")

plt.show()
```

```python
bar_df = df.groupby('Species').max()
bar_df
```

```python
bar_df.values
```

```python
plt.figure(figsize=(10, 6))
plt.bar(bar_df.index, bar_df['SepalLengthCm'], color = ['r', 'b', 'y'], edgecolor = 'k')
plt.title("Maximum Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Maximum Sepal Length (cm)")

plt.show()
```

### D. Visualizing two-dimensional data using heatmaps


```python
heat_df = df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
heat_df.corr()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(heat_df.corr(), annot=True, cmap='coolwarm', vmin = -1, vmax = 1)
plt.title("Correlation Heatmap")
plt.show()
```

### F. Plotting multiple Matplotlib and Seaborn charts in a grid


```python
df
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

axes[0, 0].scatter(df['SepalLengthCm'], df['SepalWidthCm'], color='blue', alpha=0.7)
axes[0, 0].set_title('Sepal Length vs Sepal Width')
axes[0, 0].set_xlabel('Sepal Length')
axes[0, 0].set_ylabel('Sepal Width')

sns.histplot(df['SepalLengthCm'], bins=20, ax=axes[0, 1], color='green')
axes[0, 1].set_title('Petal Length Distribution')

sns.boxplot(x='Species', y='SepalWidthCm', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Sepal Width by Species')

sns.heatmap(heat_df.corr(), annot=True, cmap='coolwarm', ax=axes[1, 1])
axes[1, 1].set_title('Correlation Heatmap')

plt.tight_layout()
plt.show()
```
